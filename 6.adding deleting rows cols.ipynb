```python
import pandas as pd
people = {
    'first': ['Corey', 'Jane', 'John'],
    'last': ['Schafer', 'Doe', 'Doe'],
    'email': ['CMS@gmail.com', 'JaneDoe@gmail.com','JD@gmail.com']
}

dftest = pd.DataFrame(people)
```

```python
dftest['full_name'] = dftest['first'] + ' ' + dftest['last']
```

```python
dftest
```

```python
dftest.drop(columns = ['last', 'first'], inplace = True)
dftest
```

```python
dftest['full_name'].str.split(' ', expand = True)
```

```python
dftest[['first', 'last']] = dftest['full_name'].str.split(' ', expand = True)
dftest
```

```python
dftest.append({'first': 'Tony'}, ignore_index = True)
```
